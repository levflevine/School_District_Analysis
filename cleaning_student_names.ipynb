```python
import pandas as pd
import os
```

```python
file_to_open = os.path.join ("Resources", "students_complete.csv")
student_data_df = pd.read_csv(file_to_open)
```

```python
# Put the student names in a list.
student_names = student_data_df["student_name"].tolist()
#student_names
```

```python
student_wrong_names=[]
for rr in student_names:
    ee = rr.split()
    tt = len(ee)
    if tt >= 3:
        student_wrong_names.append(rr)
        #print(f'Names: {ee}. Length: {tt}')
        #len (student_wrong_names)
```

```python
short_prefix = []
for yy in student_wrong_names:
    if len(yy.split()[0])<=4:
        short_prefix.append(yy.split()[0])
short_prefix
```

```python
short_suff = []
for yy in student_wrong_names:
    if len(yy.split()[-1])<=3:
        short_suff.append(yy.split()[-1])
short_suff
```
